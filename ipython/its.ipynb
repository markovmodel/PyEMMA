Helper functions
-----

```python
def discretize(_trajs, _bounds):
    """
    discretizes the trajectories into two states split at point p
    """
    _dtrajs = []
    for _traj in _trajs:
        _dtraj = np.digitize(_traj,_bounds)
        _dtrajs.append(_dtraj)
    return _dtrajs   
```

```python
def determine_lengths(dtrajs):
    """
    Determines the lengths of all trajectories
    """
    if (isinstance(dtrajs[0],(int,int64))):
        return len(dtrajs)*np.ones((1))
    lengths = np.zeros((len(dtrajs)))
    for i in range(len(dtrajs)):
        lengths[i] = len(dtrajs[i])
    return lengths
```

```python
def number_of_states(dtrajs):
    # determine number of states n
    nmax = 0
    for dtraj in dtrajs:
        nmax = max(nmax, np.max(dtraj))
    # return number of states
    return nmax + 1
```

```python
def plot_implied_timescales(ITS, xlog=False, ylog=True, confidence=0.95, refs=None):
    """
    Pretty implied timescale plot
    """
    colors = ['blue','red','green','cyan','purple','orange','violet']
    lags = ITS.get_lagtimes()
    xmax = np.max(lags)
    ymin = min(np.min(lags), np.min(ITS.get_timescales()))
    for i in range(ITS.number_of_timescales()):
        # plot estimate
        plot(lags, ITS.get_timescales(i), color = colors[i % len(colors)])
        # sample available?
        if (ITS.samples_available()):
            # plot sample mean
            plot(lags, ITS.get_sample_mean(i), marker='o', color = colors[i])
            (lconf, rconf) = ITS.get_sample_conf(0.95, i)
            fill_between(lags, lconf, rconf, alpha=0.2, color=colors[i])
        # reference available?
        if (refs != None):
            tref = refs[i]
            plot([0,min(tref,xmax)], [tref,tref], color='black', linewidth=1)
    # cutoff
    fill_between(lags, ymin*np.ones(len(lags)), lags, alpha=0.5, color='grey')
    plot(lags, lags, linewidth=2, color='black')
    xlim(1,xmax)
    # formatting
    xlabel('lag time (steps)')
    ylabel('timescale (steps)')
    if (xlog):
        semilogy()
    if (ylog):
        semilogy()
```

```python
lags = [1,2,5,10,15,20,25,30,35,40]
# compute implied timescales
ITS = ImpliedTimescales(dtrajs_1, lags = lags, nits = 5, connected = True, reversible = True)
# bootstrap
ITS.bootstrap(50)
```

```python
#plot
plot_implied_timescales(ITS)
ylim(10,1000)
```
